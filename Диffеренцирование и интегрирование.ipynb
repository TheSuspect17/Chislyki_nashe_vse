```python
def f(x):
    return np.exp(x)
x = np.linspace(0, 10, 100)
```

```python
Y = f(x)
```

1 задача, диffеренцирование

```python
def diff(Y,X,N):
    S = 0.0
    y2 = []
    t = np.diff(Y)
    for i in range(99):
        S += 0.5 * (Y[i+1] - Y[i - 1]) / (X[i+1] - X[i-1])
        y2.append(S)
    plt.plot(X,Y)
    plt.plot(X[0:99],t)
    with open('diff.csv', 'w', newline='') as csvfile:
        filewriter = csv.writer(csvfile, delimiter=',',
        quotechar='|', quoting=csv.QUOTE_MINIMAL)
        filewriter.writerow(['x', 'f встроенная', 'f наше', 'разница'])
        for i in range(99):
            filewriter.writerow([X[i],t[i],y2[i],y2[i]-t[i]])
```
